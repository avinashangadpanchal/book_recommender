## Popularity based Recommendation System

```python
ratings_with_names = ratings.merge(books, on= "ISBN")
```

```python
no_ratings_df = ratings_with_names.groupby('Book-Title').count()['Book-Rating'].reset_index()
no_ratings_df.rename(columns= {'Book-Rating': 'no_ratings'},inplace = True)
no_ratings_df
```

```python
avg_ratings_df = ratings_with_names.groupby('Book-Title').mean(numeric_only = True)['Book-Rating'].reset_index()
avg_ratings_df.rename(columns= {'Book-Rating': 'average_ratings'},inplace = True)
avg_ratings_df
```

```python
popular_df= no_ratings_df.merge(avg_ratings_df, on = "Book-Title")
popular_df
```

```python
#sort the values with no of ratings above 250, and filter out 50 books with highest avg rating
popular_df = popular_df[popular_df['no_ratings']>=250].sort_values('average_ratings',ascending = False).head(50)
popular_df
```

```python
popular_df = popular_df.merge(books, on = "Book-Title").drop_duplicates('Book-Title')[["Book-Title","Book-Author","Image-URL-M","no_ratings", "average_ratings"]]
```

## Collabarative filtering base recommendation system

```python
x = ratings_with_names.groupby('User-ID').count()['Book-Rating']>200
top_readers = x[x].index
top_readers
```

```python
filtered_rating = ratings_with_names[ratings_with_names['User-ID'].isin(top_readers)]
filtered_rating
```

```python
y = filtered_rating.groupby("Book-Title").count()["Book-Rating"]>=50
famous_books = y[y].index
famous_books
```

```python
final_rating = filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
final_rating
```

```python
pt = final_rating.pivot_table(index='Book-Title', columns ='User-ID', values = 'Book-Rating')
pt.fillna(0, inplace= True)
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_score = cosine_similarity(pt)
```

```python
similarity_score
```

```python
def recommend(book_name):
    #fetch index
    index = np.where(pt.index == book_name)[0][0]
    similar_items = sorted(list(enumerate(similarity_score[index])), key = lambda x:x[1], reverse = True)[1:6]
    
    for i in similar_items:
        print(pt.index[i[0]])
```

```python
recommend('The Notebook')
```
